```python
import dolfin as df
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
import meshio
%matplotlib inline
```

```python
# mesh
Lx = 4.0
Ly = 4.0
a = 0.3
dx = 0.03
N = 55
seed = 5

mesh = df.Mesh()
with df.XDMFFile("4 July/periodic_porous_Lx{}_Ly{}_a{}_N{}_dx{}_seed{}_xdmf.xdmf".format(Lx, Ly, a, N, dx, seed)) as infile:
    infile.read(mesh)
    
df.plot(mesh, title="Mesh")
plt.show()
```

```python
class Left(df.SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and df.near(x[0], -Lx/2)

class Right(df.SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and df.near(x[0], Lx/2)

class Bottom(df.SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and df.near(x[1], -Ly/2)

class Top(df.SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and df.near(x[1], Ly/2)

class Grains (df.SubDomain):
    def inside(self, x, on_boundary):
        return on_boundary and x[0] > -Lx/2 + df.DOLFIN_EPS and x[0] < Lx/2 - df.DOLFIN_EPS \
                and x[1] > -Ly/2 + df.DOLFIN_EPS and x[1] < Ly/2 - df.DOLFIN_EPS
    
left = Left()
top = Top()
right = Right()
bottom = Bottom()
grains = Grains()
```

```python
mark = {"generic": 0,
"left": 1,
"top": 2,
"right": 3,
"bottom":4,
"grains":5}

subdomains = df.MeshFunction("size_t", mesh, 1)
subdomains.set_all(mark["generic"])
left.mark(subdomains, mark["left"])
top.mark(subdomains, mark["top"])
right.mark(subdomains, mark["right"])
bottom.mark(subdomains, mark["bottom"])
grains.mark(subdomains, mark["grains"])
```

```python
# Periodic boundary condition

class PeriodicDomain(df.SubDomain):

    def inside(self, x, on_boundary):
        # return True if on left or bottom boundary AND NOT on one of the two slave edges
        return bool((df.near(x[0], -Lx/2) or df.near(x[1], -Ly/2.)) and 
            (not ((df.near(x[0], Lx/2) and df.near(x[1], -Ly/2.)) or 
                  (df.near(x[0], -Lx/2) and df.near(x[1], Ly/2.)))) and on_boundary)

    def map(self, x, y):
        if df.near(x[0], Lx/2) and df.near(x[1], Ly/2.):
            y[0] = x[0] - Lx
            y[1] = x[1] - Ly
        elif df.near(x[0], Lx/2):
            y[0] = x[0] - Lx
            y[1] = x[1]
        elif df.near(x[1], Ly/2.):
            y[0] = x[0]
            y[1] = x[1] - Ly
        else:
            y[0] = -1000
            y[1] = -1000
            
pbc = PeriodicDomain()
```

```python
# Define variational problem

V = df.VectorElement("CG", mesh.ufl_cell(), 2)
P = df.FiniteElement("CG", mesh.ufl_cell(), 1)
VP = df.MixedElement([V, P])
W = df.FunctionSpace(mesh , VP, constrained_domain = pbc)

(u, p) = df.TrialFunctions(W)
(v, q) = df.TestFunctions(W)
```

```python
# No slip boundary condition

#bc_left = df.DirichletBC(W.sub(0),df.Constant((0.0, 0.0)), subdomains, mark["left"])
#bc_right = df.DirichletBC(W.sub(0),df.Constant((0.0, 0.0)), subdomains, mark["right"])
bc_grains = df.DirichletBC(W.sub(0),df.Constant((0.0, 0.0)), subdomains, mark["grains"])

bcs = [bc_grains]
```

```python
# Volume integration
dx = df.Measure("dx", domain=mesh, subdomain_data=subdomains)

# Surface integration
ds = df.Measure("ds", domain=mesh, subdomain_data=subdomains)
```

```python
# Body force:
force = df.Constant((0.0, 1.0))
```

```python
# Surface normal
n = df.FacetNormal(mesh)
```

```python
a = df.inner(df.grad(u), df.grad(v))*dx - p*df.div(v)*dx + q*df.div(u)*dx
L = df.inner(force, v) * dx
```

```python
w = df.Function(W)

# Compute solution
df.solve(a == L, w, bcs)
```

```python
(u, p) = w.split(True)
```

```python
fig1 = df.plot(u, title = "Velocity")
plt.show()
```

```python
fig2 = df.plot(p,title="Pressure")
plt.show()
```

```python
# Save the solution to open in ParaView.

ufile = df.File("velocity/velocity_N{}_seed{}.pvd".format(N, seed))
ufile << u
pfile = df.File("pressure/pressure_N{}_seed{}.pvd".format(N, seed))
pfile << p
```

```python
velocity_data = []
for i in range(1000000):
    xi = np.random.uniform(-Lx/2,Lx/2)
    yi = np.random.uniform(-Ly/2,Ly/2)
    x = df.Point(xi, yi)
    
    # This return a very large number, if out of the mesh.
    if mesh.bounding_box_tree().compute_first_entity_collision(x) < 4E9:
        u_i = u(x)
        velocity_data.append(u_i)
```

```python
speed_data=[]
for i in velocity_data:
    speed = np.sqrt(i[0]**2 + i[1]**2)
    speed_data.append(speed)
```

```python
import pickle
file_name = "speed/speed_N{}_seed{}.pkl".format(N, seed)
open_file = open(file_name, "wb")
pickle.dump(speed_data, open_file)
open_file.close()
```
