```python
import pickle

N = 55

speed_list=[]

for seed in [1,2,3,4,5]:
    with open("speed/speed_N{}_seed{}.pkl".format(N,seed), "rb") as fp:
        speed = pickle.load(fp)
        speed_list.append(speed)
```

```python
import matplotlib.pyplot as plt

n_list = []
bins_list = []

for speed in speed_list:
    n, bins, patches = plt.hist(speed, bins=500)
    n_list.append(n)
    bins_list.append(bins)

```

```python
plt.title("{} obstacles".format(N))
plt.xlabel('Speed v')
plt.ylabel('PDF p(v)')
plt.style.use('classic')

pdf = plt.plot(bins_list[0][0:500],n_list[0], bins_list[1][0:500],n_list[1], 
               bins_list[2][0:500],n_list[2], bins_list[3][0:500],n_list[3], bins_list[4][0:500],n_list[4])

plt.legend(pdf, ["Case 1","Case 2","Case 3","Case 4","Case 5"], shadow=True)

plt.savefig("speed/Speed PDF _ N{}.png".format(N), dpi=300)
plt.show()
```
